```python
import json
import pandas as pd

with open('/Users/tyler/Downloads/liked_posts-1.json', 'r') as f:
    data = json.load(f)

likes_list = data['likes_media_likes']

rows = []
for entry in likes_list:
    account = entry['title']
    for item in entry['string_list_data']:
        link = item.get('href', '')
        like_value = item.get('value', '')
        timestamp = item.get('timestamp')
        rows.append([account, link, like_value, timestamp])

df = pd.DataFrame(rows, columns=['account', 'link', 'like', 'timestamp'])

df.head()
```

```python
df['datetime'] = pd.to_datetime(df['timestamp'], unit='s')

df[['account', 'datetime']].head()
```

```python
like_counts = df.groupby('account').size().reset_index(name='count')

like_counts = like_counts.sort_values(by='count', ascending=False).reset_index(drop=True)

like_counts.head()
```

```python
with open('/Users/tyler/Downloads/liked_posts-1.json','r') as f:
    liked_data = json.load(f)

likes_list = []
for item in liked_data["likes_media_likes"]:
    account = item["title"]
    for event in item["string_list_data"]:
        likes_list.append({
            'account': account,
            'href': event.get('href'),
            'timestamp': event.get('timestamp')
        })

df = pd.DataFrame(likes_list)
```

```python
df['datetime'] = pd.to_datetime(df['timestamp'], unit='s')

df['day_of_week'] = df['datetime'].dt.day_name()

df.head(10)
```

```python
likes_by_day = df['day_of_week'].value_counts()
print(likes_by_day)
```
